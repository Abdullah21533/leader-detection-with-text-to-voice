```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import spacy
from gtts import gTTS
import os

# Load SpaCy model
nlp = spacy.load('en_core_web_sm')

# Initialize NLTK sentiment analyzer
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()

# Define leadership traits and keywords
leadership_keywords = {
    'assertive': ['decisive', 'confident', 'clear', 'strong', 'authoritative'],
    'visionary': ['future', 'strategy', 'direction', 'goal', 'plan'],
    'responsibility': ['accountable', 'responsible', 'own', 'commitment'],
    'communication': ['clarity', 'communicate', 'message', 'share', 'information']
}

def analyze_text(text):
    doc = nlp(text)
    score = 0

    # Check for presence of leadership traits based on keywords
    for trait, keywords in leadership_keywords.items():
        if any(keyword in text.lower() for keyword in keywords):
            score += 1

    # Analyze sentiment
    sentiment_score = sia.polarity_scores(text)['compound']
    if sentiment_score > 0.2:
        score += 1  # Positive sentiment can also be a sign of leadership

    return score

def text_to_speech(text, filename):
    tts = gTTS(text, lang='en')
    tts.save(filename)

def process_file(file_path, output_prefix):
    with open(file_path, 'r') as file:
        texts = file.read().split('\n\n')

    for i, text in enumerate(texts):
        score = analyze_text(text)
        if score >= 2:  # Adjust threshold based on your needs
            print(f"Leadership Detected in Text {i+1} from {file_path}")
            text_to_speech(text, f'{output_prefix}_text_{i+1}.mp3')
        else:
            print(f"No Leadership Detected in Text {i+1} from {file_path}")

# Process leadership texts
process_file('leadership_texts.txt', 'leadership')

# Process non-leadership texts
process_file('non_leadership_texts.txt', 'non_leadership')
```
